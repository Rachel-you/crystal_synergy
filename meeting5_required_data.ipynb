### Create dataframe with year as filter, column with 1-12 months, non_month_columns 'C1AccountNo', 'Comp Terr', 'Latitude'

```python
non_month_columns = ['C1AccountNo', 'CXRecords', 'Comp Terr', 'Latitude']

yearly_data = {}

for year in range(2018, 2025):
    year_columns = [col for col in df_without_sale_q.columns if str(year) in col]

    df_year = df_without_sale_q[non_month_columns + year_columns].copy()
    month_mapping = {f'{year}_{str(i).zfill(2)}': i for i in range(1, 13)}
    df_year.rename(columns=month_mapping, inplace=True)

    df_year['Year'] = year

    yearly_data[year] = df_year

df_concat = pd.concat([yearly_data[2018], yearly_data[2019], yearly_data[2020],
                      yearly_data[2021], yearly_data[2022], yearly_data[2023], yearly_data[2024]] ,ignore_index=True)
```

```python
import pandas as pd
new_df_concat = pd.DataFrame()
for year in range(2018, 2025):
  df_year = df_concat[df_concat['Year'] == year].copy()
  month_columns = [col for col in df_year.columns if isinstance(col, int) and 1 <= col <= 12]
  df_year_deduplicated = df_year.drop_duplicates(subset=month_columns, keep='first')
  new_df_concat = pd.concat([new_df_concat, df_year_deduplicated], ignore_index=True)


print(len(new_df_concat[new_df_concat['Year'] == 2024]))
new_df_concat.head()
len(new_df_concat)
```

```python
import pandas as pd

# Create an empty list to hold yearly dataframes
yearly_dfs = []

# Loop through the years and select relevant columns for each
for year in range(2018, 2024):  # Up to 2023 only
    yearly_data = {
        'C1AccountNo': df_sale_q['C1AccountNo'],
        'CXRecords': df_sale_q['CXRecords'],
        'q1': df_sale_q[f'{year}_q1'],
        'q2': df_sale_q[f'{year}_q2'],
        'q3': df_sale_q.get(f'{year}_q3', pd.NA),  # Handle missing columns
        'q4': df_sale_q.get(f'{year}_q4', pd.NA),
        'sales': df_sale_q[f'{year}_sales'],
        'individual': df_sale_q.get(['individual'],pd.NA),  # Including the 'individual' column
        'Year': year
    }
    yearly_df = pd.DataFrame(yearly_data)
    yearly_dfs.append(yearly_df)

# For 2024, handle only q1, q2, and sales
year_2024_data = {
    'C1AccountNo': df_sale_q['C1AccountNo'],
    'CXRecords': df_sale_q['CXRecords'],
    'q1': df_sale_q['2024_q1'],
    'q2': df_sale_q['2024_q2'],
    'q3': pd.NA,  # Missing q3
    'q4': pd.NA,  # Missing q4
    'sales': df_sale_q['2024_sales'],
    'individual': df_sale_q.get(['individual'],pd.NA),  # Including the 'individual' column
    'Year': 2024
}
yearly_df_2024 = pd.DataFrame(year_2024_data)
yearly_dfs.append(yearly_df_2024)
```

```python
import pandas as pd
# Concatenate all yearly dataframes into a single dataframe
df_yearly_sales = pd.concat(yearly_dfs, ignore_index=True)
df_yearly_sales.head()
```

```python
df_yearly_sales = df_yearly_sales.fillna(0)
df_yearly_sales.head()
```

```python
merged_df = pd.merge(new_df_concat, df_yearly_sales, on=['C1AccountNo','CXRecords','Year'])
merged_df.head()
```

```python
# prompt: Calculate the quarterly sales for each C1AccountNo based on the monthly sales for that quarter ,give new df,

# Calculate quarterly sales
def calculate_quarterly_sales(row):
    year = row['Year']
    q1_sales = 0
    q2_sales = 0
    q3_sales = 0
    q4_sales = 0
    sales = 0

    for month in range(1, 13):
        month_col_name = month
        if month_col_name in row:
          if month in [1, 2, 3]:
              q1_sales += row[month_col_name]
          elif month in [4, 5, 6]:
              q2_sales += row[month_col_name]
          elif month in [7, 8, 9]:
              q3_sales += row[month_col_name]
          elif month in [10, 11, 12]:
              q4_sales += row[month_col_name]

    total_sales = sum([q1_sales, q2_sales, q3_sales, q4_sales])
    return q1_sales, q2_sales, q3_sales, q4_sales, total_sales


merged_df['q1_calculated'], merged_df['q2_calculated'], merged_df['q3_calculated'], merged_df['q4_calculated'], merged_df['individual_calculated'] = zip(*merged_df.apply(calculate_quarterly_sales, axis=1))


# Display the new dataframe with calculated quarterly sales
merged_df.head()
```
