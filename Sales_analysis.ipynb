```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
df = pd.read_csv('cleaned_data_V3.csv')
```

```python
df
```

### **Quarterlt sale plot**

```python
br1 = df[['Year','q1_sales']].groupby(['Year']).sum()['q1_sales']
br2 = df[['Year','q2_sales']].groupby(['Year']).sum()['q2_sales']
br3 = df[['Year','q3_sales']].groupby(['Year']).sum()['q3_sales']
br4 = df[['Year','q4_sales']].groupby(['Year']).sum()['q4_sales']
x = df['Year'].unique()
```

```python
X_axis = np.arange(len(x))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
ax1.bar(X_axis, br1, 0.2, label = 'q1_sales')
ax1.bar(X_axis+0.2, br2, 0.2, label = 'q2_sales')
ax1.bar(X_axis+0.4, br3, 0.2, label = 'q3_sales')
ax1.bar(X_axis+0.6, br4, 0.2, label = 'q4_sales')

ax1.set_xticks(X_axis)
ax1.set_xticklabels(x)
ax1.set_xlabel("Year")
ax1.set_ylabel("quarterly sales")
ax1.set_title("quarterly sales bar chart")
ax1.legend()

ax2.plot(x, br1, label = 'q1_sales')
ax2.plot(x, br2, label = 'q2_sales')
ax2.plot(x, br3, label = 'q3_sales')
ax2.plot(x, br4, label = 'q4_sales')
for year, value in br1.items():
    ax2.text(year, value, f'{value:.1f}', ha='center', va='bottom')
for year, value in br2.items():
    ax2.text(year, value, f'{value:.1f}', ha='center', va='bottom')
for year, value in br3.items():
    ax2.text(year, value, f'{value:.1f}', ha='center', va='bottom')
for year, value in br4.items():
    ax2.text(year, value, f'{value:.1f}', ha='center', va='bottom')

ax2.set_xlabel("Year")
ax2.set_ylabel("quarterly sales")
ax2.set_title("quarterly sales line chart")
ax2.legend()
plt.show()
```

### **Sales ditribution by region (able to interaction)**

```python
region_sales = df[['Year','Territory','annual_sales']].groupby(['Year','Territory']).sum()
```

```python
region_sales_unstacked = region_sales.unstack()
```

```python
region_sales_unstacked.index
```

```python
region_sales_unstacked.columns
```

```python
import plotly.graph_objects as go

# Reshape the data for plotting
region_sales_unstacked = region_sales.unstack()

# Create the Plotly figure
fig_bar = go.Figure()

for territory in region_sales_unstacked.columns.get_level_values(1).unique():
  fig_bar.add_trace(go.Bar(
      x=region_sales_unstacked.index,
      y=region_sales_unstacked[('annual_sales', territory)],
      name=territory
  ))

fig_bar.update_layout(
    title='Sales by Year and Territory',
    xaxis_title='Year',
    yaxis_title='Annual Sales',
    hovermode="x unified",
    width=1400,
    height=600
)

fig_bar.show()


fig_line = go.Figure()
for territory in region_sales_unstacked.columns.get_level_values(1).unique():
  fig_line.add_trace(go.Scatter(
      x=region_sales_unstacked.index,
      y=region_sales_unstacked[('annual_sales', territory)],
      mode='lines',
      name=territory
  ))

fig_line.update_layout(
    title='Sales by Year and Territory',
    xaxis_title='Year',
    yaxis_title='Annual Sales',
    hovermode="x unified",
    width=1400,
    height=600
)

fig_line.show()
```

```python
# Define bins for annual sales classification
bins = [0, 1500, 5000, 10000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

df['sales_class'] = pd.cut(df['annual_sales'], bins = bins, labels=labels)

df[['Year', 'annual_sales', 'sales_class']]
```

```python
print('Mention the classification here:')
print('sales class rank: 0-1000 are low, 1000-5000 are medium, 5000-10000 are high, > 10000 are very high, divide by distribution plot above.')

# Bar plot for sales_class
warnings.filterwarnings("ignore")
sales_class_counts = df.groupby(['Year', 'sales_class'])['annual_sales'].count().unstack()
sales_class_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Sales Class Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Number of Sales')
plt.legend(title='Sales Class')
plt.show()
```
